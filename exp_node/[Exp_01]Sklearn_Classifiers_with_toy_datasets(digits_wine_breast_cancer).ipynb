scikit-learn에서 여러데이터셋 및 머신러닝 모델에 대하여 동일한 방식으로 데이터셋을 로드하고, 학습을 진행하므로 반복 작업에 대하여 편의를 제공하고, python에 대한 이해를 높이기 위해 SciKitModelTester를 작성함.

실제 데이터셋에서는 데이터에 대한 전처리가 필요하겠지만 여기서는 주어진 데이터셋을 그대로 이용하여 학습을 진행함.


프로젝트 요구사항.
(1) 필요한 모듈 import하기
(2) 데이터 준비
(3) 데이터 이해하기
(4) train, test 데이터 분리
(5) 다양한 모델로 학습시켜보기
(6) 모델을 평가해 보기

아래 테스트 코드에서 main을 실행하면 

1. 각각의 데이터셋으로 로드하고,
2. train, test데이터 분리.
3. 설정한 모델들에 대하여 학습을하고, 
4. scikit에서 제공한 report를 실행한다.
5. 테스트별 summary를 표시하고,
6. 각 데이터셋에 대한 모델별 report를 그래프로 표시함. 

위 학습으로 아래와 같이 데이터셋별 가장 좋은 모델을 얻음.
  1. digits(64, 10) : svm.SVC모델.
  2. wine(13 * 3) : RandomForestClassfier모델
  3. breast cancel(30 * 2) : RandomForestClassfier 모델

데이터셋의 특성에 따라, 크게 4가지 모델로 분류되지만, 코드상의 주석처럼 모델별(Estimator) 설정해야할 내부 알고리즘이 여러가지이고, 이에 따른 파라메타가 여러개이기 때문에 각 모델별 특성에 대한 이해를 통하여 최적화가 가능할것 같습니다.

또한 데이터셋에 대한 전처리가 없었는데, 데이터에 대한 이해 및 Python pandas나 ndarray 이용한 데이터 처리에  숙달되어야 할 필요성을 많이 느꼈습니다.

마지막으로 그래프 라이브러리 사용 및 python언어도 (Singleton이라던가, 데이터가 클 경우에 처리방법등) 숙달이 되어야 원만한 테스트 진행이 될것 같습니다.

```python
# [Exp-01] Sklearn_Classifiers_with_toy_datasets(digits_wine_breast_cancer)

import io
import copy

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# for dataset
from sklearn.datasets import load_iris, load_digits, load_wine, load_breast_cancer

# for split dataset
from sklearn.model_selection import train_test_split

# for model
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# for report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


#scikit-learn requires:
# Python (>= 3.8)
# NumPy (>= 1.17.3)
# SciPy (>= 1.3.2)
# joblib (>= 1.0.0) 
# threadpoolctl (>= 2.0.0)

# model category(classification, regression, clustering, dimensionality reduction)
# scikit-model tester : 
# prepare Data
# ETL(Categorical Expansion, Null Imputation, Feature Scaling) : Transformer class
# Train, :  model  input: FeatureMatrix(X), output: TargetVector(Y)
# Validate, 
# Predict, : estimator predict
# Report

class SciKitModelTester:
    """SciKitModelTester 
        sckkit learn model test class
    Params:
        config(데이터셋, 모델명 및 기타 테스트관련 설정.)
    Raises:
        config check exceptions.

    Returns:
        None
    """

    # scikit datasets
    _loaders = {"iris": load_iris,
                "digits": load_digits,
                "wine": load_wine,
                "breast cancel": load_breast_cancer }

    # scikit models
    _supported_models = {"DecisionTreeClassifier": DecisionTreeClassifier,
                "RandomForestClassifier": RandomForestClassifier,
                "svm.SVC": svm.SVC,
                "SGDClassifier": SGDClassifier,
                "LogisticRegression": LogisticRegression }

    # scikit model_categorys
    _models_categorys = {"DecisionTreeClassifier": "classification",
                         # params: criterion='gini', splitter='best', max_depth=None, min_samples_split=2, 
                         #          min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, 
                         #          random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, 
                         #          class_weight=None, ccp_alpha=0.0
                         # check: get_depth()

                        "RandomForestClassifier": "classification", # meta..
                         # params: n_estimators=100, *, criterion='gini', max_depth=None, min_samples_split=2, 
                         #          min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', 
                         #          max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, 
                         #          n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, 
                         #          ccp_alpha=0.0, max_samples=None
                         # check: 

                        "svm.SVC": "Classification", # Support Vector Machines, C-Support Vector Classification
                         # params: C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, shrinking=True, 
                         #          probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, 
                         #          max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None
                         # check: decision_function get_params
                         # samle수가 많으면 안좋음.

                        "SGDClassifier": "Classification",
                         # stochastic gradient descent (SGD), partial_fit
                         # params: loss='hinge', *, penalty='l2', alpha=0.0001, l1_ratio=0.15, fit_intercept=True, 
                         #          max_iter=1000, tol=0.001, shuffle=True, verbose=0, epsilon=0.1, n_jobs=None, 
                         #          random_state=None, learning_rate='optimal', eta0=0.0, power_t=0.5, early_stopping=False, 
                         #          validation_fraction=0.1, n_iter_no_change=5, class_weight=None, warm_start=False, average=False
                         # check: 
                         
                        "LogisticRegression": "Regression"
                         # params: penalty='l2', *, dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, 
                         #                  class_weight=None, random_state=None, solver='lbfgs', max_iter=100, 
                         #                  multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None
                         # check: 
                        }
                         

    # default conifg for test.
    _default_config = {"dataset": "iris",
                       "models" : ["DecisionTreeClassifier"],
                       "model" : "DecisionTreeClassifier",
                       "run_model" : True,
                       "split_size" : 0.2,
                       "split_random" : 7,
                       "model_random" : 32,
                       "zero_division" : 'warn',
                       "show_summary" : False,
                       }


    def __init__(self):
        """ 
          주요하게 사용하는 변수들만 표시
          _summary는 io이므로 test함수 내에서만 열고 닫는다.
        """
        self._config = None
        self._dataset = None
        self._dataset_name = ""
        self._model_name = None
        self._models = []
        self._results = {}
        self._result = {'accuracy': [],
                        'accuracy': [],
                        'precision': [],
                        'recall': [], 
                        'f1-score': [] }

    
    @classmethod
    def get_config(cls):
        """ 
          default_config를 새로 생성해서 넘겨줌.
        """      
        return copy.deepcopy(cls._default_config);

    
    def config_val(self, name):
        """ 
          config에 필요한 항목값을 가져오는 함수
          해당 항목이 없으면 Exception을 발생.
        """
        if name not in self._config:
            raise Exception("config [%s] not found" % name)

        return self._config[name]


    def check_config(self, config):
        """ 
          config에서 가장 중요한 dataset, model 항목 여부만 체크함.
        """

        self._config = config

        if self.config_val("dataset") not in self._loaders:
            raise Exception("config dataset not found")
        
        if self.config_val("model") not in self._supported_models:
            raise Exception("config model not found")

        for model in self.config_val("models"):
            if model not in self._supported_models:
                raise Exception("config model not found")
    
    def load_dataset(self):
        """ 
          config에서 설정한 dataset을 로드하여 _dataset 에 저장함.
          데이터가 많은경우, 객체가 여러개일때 어떻게 동작하는지 확인이 필요함.
        """

        dataset_name = self.config_val("dataset")
        if dataset_name in self._loaders:
            loader = self._loaders[dataset_name]
            self._dataset = loader();
            self._dataset_name = dataset_name
        else :
            raise Exception("not supported dataset...")
        
        self._summary.write(str(dataset_name) + " dataset loaded \n")


    def dataset_info(self):
        """ 
          dataset을 확인하고 정보(데이터갯수, features, labels등)를 summary에 남김
        """

        if "data" not in self._dataset.keys():
            raise Exception("data not found")

        if 2 != self._dataset.data.ndim :
            raise Exception("data not 2dim")

        if "target" not in self._dataset.keys():
            raise Exception("target not found")

        data = pd.DataFrame(self._dataset.data, columns=self._dataset.feature_names)

        if "feature_names" in self._dataset.keys() and data.count == len(self._dataset.feature_names):
            data.rename(columns = self._dataset.feature_names, inplace = True)
        
        data.info(buf=self._summary)
        #self._summary.write("features (%d) : %s\n" %  (int(data.shape[1]),  str(data.columns.values)))
        #self._summary.write("size : %d \n" %  int(data.shape[0]))
        self._summary.write("target_names : %s \n" % self._dataset.target_names)
        #self._summary.write("DESCR : %s \n" % self._dataset.DESCR)
        self._summary.write("\n")


    def prepare_data(self):
        """ 
          데이터를 학습과 검증용으로 나눔.
        """

        split_size = self.config_val("split_size")
        split_random = self.config_val("split_random")
        self._X_train, self._X_test, self._y_train, self._y_test = \
            train_test_split(self._dataset.data, self._dataset.target, test_size=split_size, random_state=split_random)

        self._summary.write("prepare data :  X size: %d, y size: %d (ratio: %0.2f, random: %d)\n\n" % (len(self._X_train), len(self._X_test), split_size, split_random))

    def set_model(self, model_name):
        """ 
          학습 모델을 선택하.
        """

        model_random = self.config_val("model_random")
        if model_name in self._supported_models:
            model = self._supported_models[model_name]
            self._model = model(random_state=model_random)
            self._model_name = model_name
        else:
            raise Exception("not supported model...")


    def train_model(self):
        """ 
          주어진 데이터로 학습하기.
        """

        self._model.fit(self._X_train, self._y_train)


    def do_predict(self):
        """ 
          학습된 모델에 대하여 검증데이터로 결과치를 예측
        """

        self._y_pred = self._model.predict(self._X_test)


    def do_report(self):
        """ 
          정확도(Accuracy)와 
          오차행렬을 통하여 정밀도(Precision), 재현율(Recall, Sensitivity), F1 스코어(f1 score)를 확인함.
          결과값은 macro avg와 weighted macro avg중에서 선택.
        """
        
        # to check cmatrix
        cmatrix = confusion_matrix(self._y_test, self._y_pred)
        #self._summary.write("\nc-matrix :\n" + str(cmatrix) + "\n")

        zd_val = self.config_val("zero_division")
        if "wine" == self._dataset_name and "LogisticRegression" == self._model_name:
            zd_val = 'warn'

        self._report = classification_report(self._y_test, self._y_pred, output_dict=True, zero_division=zd_val)

        #print(str(self._report))

        decimal_places = 2
        accuracy = round(self._report['accuracy'], decimal_places)
        precision = round(self._report['weighted avg']['precision'], decimal_places)
        recall = round(self._report['weighted avg']['recall'], decimal_places)
        f1_score = round(self._report['weighted avg']['f1-score'], decimal_places)
        # support = self._report['weighted avg']['support']

        # accuracy = round(self._report['accuracy'], decimal_places)
        # precision = round(self._report['macro avg']['precision'], decimal_places)
        # recall = round(self._report['macro avg']['recall'], decimal_places)
        # f1_score = round(self._report['macro avg']['f1-score'], decimal_places)
        # support = self._report['macro avg']['support']

        self._result['accuracy'] = accuracy
        self._result['precision'] = precision
        self._result['recall'] = recall
        self._result['f1-score'] = f1_score

        self._summary.write("\n%25s : {" % self._model_name)
        self._summary.write("%s: %0.2f, " % ('accuracy', self._result['accuracy']))
        self._summary.write("%s: %0.2f, " % ('precision', self._result['precision']))
        self._summary.write("%s: %0.2f, " % ('recall', self._result['recall']))
        self._summary.write("%s: %0.2f, " % ('f1-score', self._result['f1-score']))
        self._summary.write("}\n ")
        
        
    def do_test(self, config):
        """ 
          config 설정값으로 선택한 데이터셋과 모델에 대하여 학습과 예측을 통하여 검증까지 일괄 진행한다.
          함수내에서 예외는 모두 처리한다.
          _summary io는 함수내에서만 open/close한다.
        """

        try:
            # open string buffer
            self._summary = io.StringIO()
            
            self.check_config(config)
            
            self.load_dataset()
            self.dataset_info()
                    
            self._results = {}  
            if self.config_val("run_model") :
                self.prepare_data()

                model_names = self.config_val("models")
                for model_name in model_names:
                    self.set_model(model_name)
                    self.train_model()
                    self.do_predict()
                    self.do_report()
                    self._results[model_name] = copy.deepcopy(self._result)
                
                if self.config_val("show_summary") :
                    # 각 테스트별 요약본 출력.
                    print("\n\n")
                    print("================================= summary report =================================\n")
                    print(self._summary.getvalue())
                    print("==================================================================================\n")

        except Exception as e:
            print("test failed : ", e)

        finally:
            # close string buffer
            self._summary.close()

            return copy.deepcopy(self._results)


if __name__ == "__main__":
    """ 
    main함수.
    """
    tester = SciKitModelTester()
    config = tester.get_config();
    config["run_model"] = True
    config["show_summary"] = True
    config["zero_division"] = 1

    # project 1,2,3용 dataset을 준비한다.
    projects = [ "digits", "wine", "breast cancel"]

    for datas in projects:
        config["dataset"] = datas
        
        model_results = {"model": [],
                        "type":[], 
                        "val":[]}

        # 학습 모델을 적용한다.
        config["models"] = ["DecisionTreeClassifier", "RandomForestClassifier",
                            "svm.SVC", "SGDClassifier", "LogisticRegression"]

        # 학습 및 검증을 실행한다.
        results = tester.do_test(config)

        # dataset에 대한 모델별 결과치를 저장.
        for model, model_result in results.items():
            for rtype, rval in model_result.items():
                model_results['model'].append(model)
                model_results['type'].append(rtype)
                model_results['val'].append(rval)

        # dataset에 대한 모델별 결과치를 차트로 표시.
        plt.style.use('default')
        rep  = pd.DataFrame(model_results)
        ax = sns.barplot(data=rep, x="model", y="val", hue="type")
        ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
        plt.title("[" + datas + "]" + " results")
        sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
        plt.show()
        

```
